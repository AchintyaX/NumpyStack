```python
# We are going to discuss joins here
# Ad we know that sometimes we need to form some custom dataset formed by joining two tables
# This is pretty common in sql

import pandas as pd
```

```python
t1 = pd.read_csv('table1.csv')
t2 = pd.read_csv('table2.csv')
```

```python
t1
# it a dataframe of users
```

```python
t2
# this is a dataframe of advertising data
```

```python
# As we can see that the tables have the user_id as a common field
```

```python
m = pd.merge(t1, t2, on='user_id') 
# the on arg allows us to find whuch column to join on
# if you provide no argument, then row index is taken as default
```
