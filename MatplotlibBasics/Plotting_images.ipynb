```python
# All images are matrix of numbers
# they are formed by taking in account the itensity of every pixel
# A(i,j) represents the pixel intensity at coordinate(i,j)
# JPEG and PNG are not matrices they are compressed 
# Decompress them to get back a matrix
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# we would be using the famous MNIStT dataset(handwritten digits)
```

```python
df = pd.read_csv("train.csv") # it takes 1 or 2 secs more than usual to load because of the size
```

```python
df.shape # check out the size
# there are 784  data points that is the number of pixels with the first column containing the labesl
```

```python
M = df.as_matrix()
```

```python
im = M[0,1:] # this means selecting the 0th row and then in that row all the columns staring from col 1
# remember that the count starts from 0
```

```python
im.shape
```

```python
# so let us understand why we have 784, so the  deal with it is that 
#  every number in the grid is a 28X28 pixel image, so in order to plot that
# we need to convert every image into 28X28 matrix
im = im.reshape(28,28)
```

```python
im.shape # so we get the desired shape
```

```python
# now lets plot it
plt.imshow(im)
```

```python
# this looks a lot like 1 so lets check to the label to see if we are correct
M[0,0] # so it is a 1
```

```python
# if you want to display the image in black and white instead of  having a heat map look
plt.imshow(im, cmap="gray")
```

```python
# let us plot the first 25 digits in the dataset
r = 5
c = 5
t = r*c
fig = plt.figure(figsize=(r,c)) # accomodating multiple images in one plot
for i in range(0,t):
    pic = M[i,1:]
    pic = pic.reshape(28,28)
    fig.add_subplot(r,c,i+1) # adds a subplot, takes 3 arguments- nrows, ncols, index
    plt.imshow(pic, cmap="binary")
plt.show()
```
