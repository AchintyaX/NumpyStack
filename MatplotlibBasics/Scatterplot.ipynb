```python
# Second type of mathematical plot

# importing all our libraries

import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# reading our data and converting it into a numpy array

A = pd.read_csv('data_1d.csv', header=None).as_matrix()
```

```python
A # checking out our data 
```

```python
x = A[:,0] # making x as all the elements in col 0
y = A[:,1] # making y as all the elements in col 1
```

```python
# scatter plots are very useful in classification problems 

plt.scatter(x,y) # plotting the scatter plot
```

```python
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("scatter-plot")
plt.scatter(x,y)
plt.show()
```

```python
x_line= np.linspace(0,100,100)
y_line = 2*x_line+1
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("plotting two curves on the same plot")
plt.plot(x_line, y_line, color="red")
plt.scatter(x,y)
plt.show()
# we can see the line chart passing throgh the points 
```
